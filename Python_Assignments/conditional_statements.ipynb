**Count how many vowels are in a given string.**

```python
string1=input('Enter the string!:')
vowels=('a','e','i','o','u')

count=0
for ch in string1:
  if ch in vowels:
    count+=1

print(f'count of vowels:{count}')
```

**Reverse a number using only a for loop (no slicing).**

```python
num=1234567
rev=0
for i in range(len(str(num))):
  digit=num%10
  rev=rev*10+digit
  num=num//10

print(rev)
```

**Find the factorial of a number.**

```python
num=int(input('Enter a number:'))
fact=1
for i in range(1,num+1):
  fact=fact*i
print(fact)
```

```python
num=int(input('Enter a num:'))
for i in range(2,num):
  if num%i==0:
    print("not prime")
    break
else:
     print("given number is prime number")
```

**Find duplicate values in a list. Example: [1,2,3,2,4,5,1]**

```python
numbers=[1,2,3,2,4,5,1]
new_list=[]

for num in numbers:
  if num not in new_list:
    new_list.append(num)
print(new_list)
```
