**Find the most frequent character in "mississippi"**

```python
print("\nQuestion 11: Find the most frequent character in 'mississippi'")
word='mississippi'
freq={}
for i in word:
  if i in freq:
    freq[i]+=1
  else:
    freq[i]=1
print(freq)
max_freq=max(freq.values())
most_frequ_char=[char for char, count in freq.items() if count==max_freq]
print(f"most frequent character: {most_frequ_char}")

```

**Question 23: Convert "hello world" to "hElLo WoRlD" (alternating case)**


```python
print("\nQuestion 23: Convert 'hello world' to 'hElLo WoRlD' (alternating case)")

string='hello world'
result=" "
for i in range(len(string)):
  if i%2==0:
    result=result+string[i].lower()
  else:
    result=result+string[i].upper()
print(result)



```

**Question 39: Find the most common word in "the quick brown fox jumps over the lazy dog"**

```python
print("\nQuestion 39: Find the most common word in 'the quick brown fox jumps over the lazy dog'")

string1='the quick brown fox jumps over the lazy dog'
word=string1.split()
count={}
for i in word:
    if i in count:
        count[i]+=1
    else:
        count[i]=1
print(count)
if max(count.values())>1:
    most_common=max(count,key=count.get)
    print(most_common)
```
