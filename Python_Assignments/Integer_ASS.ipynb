**Find the factorial of 8**

```python
num=8
fact=1
for i in range(1,num+1):
  fact=fact*i
print(fact)
```

**Calculate the average of numbers: 15, 23, 31, 42, 56**

```python
num=[15,23,31,42,56]
length=len(num)
sum=0
for i in num:
  sum=sum+i

average=sum/length
print(f"Average of {sum}/{length}:",average)
```
