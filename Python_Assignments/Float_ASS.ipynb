 **Calculate the area of a circle with radius 5.5**

```python
print("SOLVED EXAMPLE:")
print("Calculate the area of a circle with radius 5.5")
import math
radius = 5.5
area = math.pi * radius ** 2
print(f"Radius: {radius}")
print(f"Area: {area:.2f}")
print("-" * 50)
```

**Calculate the average of 3.14, 2.718, 1.618, 0.577**

```python
print("Question 1: Calculate the average of 3.14, 2.718, 1.618, 0.577")
num=[3.14,2.718, 1.618, 0.577]
length=len(num)
sum=0
for i in num:
  sum=sum+i
average=sum/length
print(f"Average of {sum}/{length}:",average)
```

**Calculate the compound interest on $1000 at 5.5% for 3 years**

```python
print("\nQuestion 3: Calculate compound interest on $1000 at 5.5% for 3 years")
 #CI=A-P # A=P*(1+R/100)^T  #principal amount (p)=1000,Annual rate(r)=5.5, Time(t)=3, Total amount(A)

P=1000
r=5.5
t=3
A=P*(1+r/100)**t
CI=A-P
print(f"Total Amount:{A}")
print(f"compound interest on $1000 at 5.5% for 3 years is:{CI}")


```

**Find the hypotenuse of a right triangle with sides 3.5 and 4.2**

```python
print("\nQuestion 4: Find the hypotenuse of a right triangle with sides 3.5 and 4.2")
import math
a=3.5
b=4.2
hyp=math.sqrt(a**2+b**2)
print(f"Hypotenuse of right triangle is:{hyp:.2f}")
```

**Calculate the volume of a sphere with radius 7.8**

```python
print("\nQuestion 5: Calculate the volume of a sphere with radius 7.8")
import math
radius=7.8
volume=(4/3)*math.pi*(radius**3)
print(f"radius:{radius}")
print(f"volume:{volume:.2f}")
```
