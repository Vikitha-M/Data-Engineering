**Exercise 5: Leap Year Checker Write a program to check if a year is a leap year: Divisible by 4 AND (not divisible by 100 OR divisible by 400)**

```python
year=int(input('Enter a year:'))
if year%4 ==0 and (year%100 !=0 or year%400 ==0):
  print("Leap Year")
else:
  print("Not Leap Year")
```

**Exercise 6: Grade Calculator with Bonus Calculate grades (A-F) but add +1 bonus grade level if attendance >90%.**

```python
marks = int(input("Enter marks: "))
attendance = int(input("Enter attendance percentage: "))

if marks >= 90:
    grade = "A"
elif marks >= 80:
    grade = "B"
elif marks >= 70:
    grade = "C"
elif marks >= 60:
    grade = "D"
else:
    grade = "F"

if attendance > 90:
    if grade == "B":
        grade = "A"
    elif grade == "C":
        grade = "B"
    elif grade == "D":
        grade = "C"
    elif grade == "F":
        grade = "D"

print("Final grand:",grade)

```

**Exercise 7: Login System with Attempts Create a login system that: Checks username and password Tracks login attempts (max 3) Locks account after failed attempts**

```python
user_admin='admin'
user_password='1234'

attempts=0
max_attempts=3

while attempts < max_attempts :
  username = input('Enter the username : ')
  password = input('Enter the password : ')

  if username == user_admin and password == user_password:
    print('Login Successful!')
    break
  else:
    attempts += 1
    print("Invalid credentials. Attempts left:", max_attempts - attempts)

if attempts == max_attempts:
    print("Account locked due to too many failed attempts.")

```

**Exercise 8: Shipping Cost Calculator Calculate shipping based on: Weight of package Destination (domestic/international) Express or standard shipping**

```python
weight = float(input("Enter package weight (kg): "))
destination = input("Enter destination (domestic/international): ").lower()
shipping_type = input("Enter shipping type (standard/express): ").lower()

# Base pricing
if destination == "domestic":
    base_cost = 5
    cost_per_kg = 2
elif destination == "international":
    base_cost = 15
    cost_per_kg = 5
else:
    print("Invalid destination")
    exit()

# Calculate cost
shipping_cost = base_cost + (weight * cost_per_kg)

# Express surcharge
if shipping_type == "express":
    shipping_cost += shipping_cost * 0.5

print("Total Shipping Cost: $", shipping_cost)
```
