Question 15: Advanced Array Operations

```python
print("\nQ15.1: Apply a function to calculate 5% interest on all account balances")
def calculate_intrest(x):
    return x*0.5
np.vectorize(calculate_intrest)(banking_data[:,1])
```

```python
print("\nQ15.2: Use np.where to categorize customers by balance ranges")
categories=np.where(banking_data[:,1]<10000,"Low Balance",
                    np.where((banking_data[:,1]>=10000)&(banking_data[:,1]<50000),"Medium Balance",
                    "High Balance"))
print(categories)
```

```python
print("\nQ15.3: Use np.select to categorize customers by credit score ranges")
conditions=[
    banking_data[:,2]<600,
    (banking_data[:,2]>=600)&(banking_data[:,2]<=750),
    banking_data[:,2]>750
]
choices=["poor credit","Average credit","Excellent credit"]
categories=np.select(conditions,choices)
print(categories)
```

```python
print("\nQ15.4: Use np.piecewise to apply different interest rates based on balance")
def low_intrest(x):
    return x*0.03
def mid_intrest(x):
    return x*0.05
def high_intrest(x):
    return x*0.07

np.piecewise(
    banking_data[:,1],
    [banking_data[:,1]<10000,(banking_data[:,1]>10000) & (banking_data[:,1]<50000), banking_data[:,1]>=50000],
    [low_intrest,mid_intrest,high_intrest]
)
```

```python
print("\nQ15.5: Use np.apply_along_axis to calculate the sum of each row")
np.apply_along_axis(np.sum,1,banking_data)
```
